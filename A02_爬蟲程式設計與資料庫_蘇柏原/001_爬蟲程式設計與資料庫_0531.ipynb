##### HW1-嘗試用request之各種呼叫功能(head,get,post,delete,patch)尋找答案
* 使用requests指令連線至[網址](http://teaching.bo-yuan.net/test/requests/)
* 運用各種呼叫方式(head, get, post, put, delete, patch )找尋答案。
* 完成後把操作過程中的所有指令保留在程式碼中，將檔案繳交上來

```python
import requests

#需要DELETE的操作
r1=requests.get('http://teaching.bo-yuan.net/test/requests/',params={
    "action":"action"
})
print(r1.text)

#記得去PUT操作
r1=requests.delete('http://teaching.bo-yuan.net/test/requests/',params={
    "action":"action",
},data={
    "id":"id"
})
print(r1.text)

#PUT完了，也要PATCH資料
r1=requests.put('http://teaching.bo-yuan.net/test/requests/',params={
    "action":"action",
},data={
    "id":"id",
    "name":"name"
})
print(r1.text)

#最後POST一筆資料吧
r1=requests.patch('http://teaching.bo-yuan.net/test/requests/',params={
    "action":"action",
},data={
    "id":"id",
    "name":"name",
    "address":"address"
})
print(r1.text)

#哈哈，答對了，請把操作過程中的所有指令保留在程式碼中，將檔案繳交上來
r1=requests.post('http://teaching.bo-yuan.net/test/requests/',params={
    "action":"action",
},data={
    "id":"id",
    "name":"name",
    "address":"address"
})
print(r1.text)
```

```python
#4-7_截圖所有圖(聯網或下載圖)

from bs4 import BeautifulSoup
import requests

r1=requests.get('https://news.ebc.net.tw/realtime',headers={
'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'
})
p1=BeautifulSoup(r1.text,'html.parser')
#抓出新聞title+summary
d1=p1.find_all('div',{'class':'text'})
fn=0
for content in d1:
  if content.find('span',{'class':'title'}) !=None:
    new_url= 'https://news.ebc.net.tw/'+ content.parent.attrs['href']
    print(new_url)
    print(content.find('span',{'class':'title'}).text)
    print(content.find('span',{'class':'summary'}).text)

    #解析各篇新聞內文
    r2=requests.get(new_url,headers={
                    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'
                    })
    p2=BeautifulSoup(r2.text,'html.parser')

    #將一行行內容儲存至news_content
    news_content=''
    fn+=1
    fn_img=0
    for d3 in p2.find('content-ad').find_all('p'):
        try:
          #下載圖片
          if d3.find('img') == None:
            news_content+=d3.text+'\r\n'
          else:
              fn_img+=1
              r3=requests.get(d3.find('img').attrs['src'],headers={
                          'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'
                          })
              with open('news/'+str(fn)+'-'+str(fn_img)+'.jpg','wb') as f:
                f.write(r3.content)
        except AttributeError as err :
          print('找不到圖片')
    with open('news/'+str(fn)+'.txt','w') as f:
      f.write(new_url+'\r\n')
      f.write(content.find('span',{'class':'title'}).text+'\r\n')
      f.write("="*30+'\r\n')
      f.write(news_content)

    print("="*30)
```

```python
import pymysql
import prettytable
import IPython.display as dp

#表格
pt=prettytable.PrettyTable(["編號","姓名","生日","地址"], encoding="utf-8")
pt.align['編號']='l'
pt.align['姓名']='l'
pt.align['生日']='l'
pt.align['地址']='l'

#登入
host_mem=input('請輸入主機位置:')
user_mem=input('請輸入資料庫帳號:')
passwd_mem=input('請輸入資料庫密碼:')
port_mem=int(input('請輸入資料的Port:'))
db_mem=input('請輸入資料名稱:')


cl=pymysql.connect(
  host=host_mem,
  user=user_mem,
  passwd=passwd_mem,
  db=db_mem,
  charset='utf8',
  port=port_mem,
)
pl=cl.cursor()
dp.clear_output(True)

def display_menu():
    print('(0) 離開程式')
    print('(1) 顯示會員列表')
    print('(2) 新增會員資料表')
    print('(3) 更新會員資料表')
    print('(4) 刪除會員資料表')

display_menu()
command=int(input('指令:'))
dp.clear_output(True)
print('')

while True:

	#1>顯示會員列表
  if command==1:
      #叫出資料庫
      pl.execute("SELECT * FROM `member`")
      #一次抓全部
      for i in pl.fetchall():
        pt.add_row([i[0],i[1],i[2],i[3]])
      print(pt)
      #轉跳至搜尋主畫面
      display_menu()
      command=int(input('指令:'))
      dp.clear_output(True)
      pt.clear_rows()
      print('')
	#2>新增會員資料表
  elif command==2:
      #SQL新增資料
      pl.execute("INSERT INTO `member`(`name`,`birthday`,`address`) VALUES(%s,%s,%s)",
          [input("請輸入會員姓名:"),input("請輸入會員生日:"),input("請輸入會員地址:")]
      )
      cl.commit()
      dp.clear_output(True)

      #轉跳至搜尋主畫面
      display_menu()
      command=int(input('指令:'))
      dp.clear_output(True)
      print('')
	#3>更新會員資料表
  elif command==3:
      #叫出資料庫
      pl.execute("SELECT * FROM `member`")
      #一次抓全部
      for i in pl.fetchall():
        pt.add_row([i[0],i[1],i[2],i[3]])
      print(pt)

      #SQL修改資料
      id_revise=input('請選擇你要修改的資料編號:')
      name_revise=input('請輸入會員姓名:')
      birthday_revise=input('請輸入會員生日:')
      address_revise=input('請輸入會員地址:')

      pl.execute(f"UPDATE `member` SET `name`='{name_revise}',`birthday`='{birthday_revise}',address='{address_revise}' WHERE `id`={id_revise}")
      cl.commit()
      dp.clear_output(True)
      pt.clear_rows()
      #轉跳至搜尋主畫面
      display_menu()
      command=int(input('指令:'))
      dp.clear_output(True)
      print('')
	#4>刪除會員資料表
  elif command==4:
      #叫出資料庫
      pl.execute("SELECT * FROM `member`")
      #一次抓全部
      for i in pl.fetchall():
        pt.add_row([i[0],i[1],i[2],i[3]])
      print(pt)
      #SQL刪除資料
      del_id=str(input('請選擇你要刪除的資料編號:'))
      pl.execute(f"DELETE FROM `member` WHERE `id`='{del_id}'")
      cl.commit()
      dp.clear_output(True)
      pt.clear_rows()
      print('')
      #轉跳至搜尋主畫面
      display_menu()
      command=int(input('指令:'))
      dp.clear_output(True)
      pt.clear_rows()
      print('')

  elif command==0:
      break
  else :
      #轉跳至搜尋主畫面
      display_menu()
      command=int(input('指令:'))
      dp.clear_output(True)
      print('')
```
