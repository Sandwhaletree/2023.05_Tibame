```python
import pandas as pd
from sklearn.datasets  import load_diabetes
diabetes=load_diabetes(scaled=False)
df=pd.DataFrame(diabetes['data'], columns=diabetes['feature_names']) #dataËΩâÁÇ∫dfÂûãÊÖãÔºåcolumnsË≥áÊñôÂÑ≤Â≠òËá≥feature_names
df['ans']=diabetes['target'] #Âä†ÂÖ•ansÊ¨Ñ‰Ωç=È≥∂Â∞æËä±Á®ÆÈ°ûtarget
df
```

```python
#Ê©üÂô®Â≠∏ÁøíÁöÑÊµÅÁ®ã‰∏çË¶ÅÁî®dfÔºåË´ãÈÉΩËΩâÊàênumpy array
import numpy as np
from sklearn.model_selection import train_test_split
x=df.drop("ans",axis=1) #dropÊéâÊ¨Ñ‰ΩçansÔºåÂè™Ââ©ÁâπÂæµ
y=df['ans']
x=np.array(x) #ÂéüÂßãÁâπÂæµÂÄº #Â∞áË≥áÊñôÂæû DFËΩâÊèõÁÇ∫ NumPy arrayÔºåÂ§öÊï∏Ê©üÂô®Â≠∏ÁøíÂ∫´Â∫ïÂ±§Êé°NumPy array
y=np.array(y) #ÂéüÂßãËä±Á®ÆÈ°û
# [x part1(90%), x part2, y partl, y part2]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1) #train_test_split ÂáΩÊï∏Â∞áË≥áÊñôÈõÜÂäÉÂàÜÁÇ∫Ë®ìÁ∑¥ÈõÜÂíåÊ∏¨Ë©¶ÈõÜ #test_size=0.1Ë°®Á§∫Ê∏¨Ë©¶ÈõÜÁöÑÊØî‰æãÊòØÊï¥ÂÄãË≥áÊñôÈõÜÁöÑ10ÔºÖ
print(x_train.shape) #Ëº∏ÂÖ•
print(x_test.shape) #Ëº∏Âá∫
print(y_train.shape) #Ëº∏ÂÖ• #(135,)=tupleÔºå‰ª£Ë°®135ÂÄãÂÖÉÁ¥†‰∏î1ÁÇ∫ÂÖÉÁ¥†ÔºåËã•Ê≤íÈÄóËôüÔºåÂ∞áÊúÉË™çÊàêÊï∏Â≠ó
print(y_test.shape) #Ëº∏Âá∫
```

```python
#ÂõûÊ≠∏DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor(max_depth=4)
# fit:Ë®ìÁ∑¥
reg.fit(x_train, y_train)
```

$ a=\\frac{1}{n}\\sum{(pre_i - true_i)}^2$  
$ b=\\frac{1}{n}\\sum{(\\mu - true_i)}^2$
$ r^2 score = 1 - \\frac{a}{b}$
0: Ë∑üÁûéÁåú(ÁåúÂπ≥Âùá)Ê≤íÂÖ©Ê®£
1: Ë∂ÖÊ∫ñ$

```python
#Áπ™Ë£Ω‰∏¶È°ØÁ§∫‰∏ÄÂÄãÊ±∫Á≠ñÊ®πÁöÑÂúñÂΩ¢ÂåñË°®Á§∫
import matplotlib.pyplot as plt #MatplotlibÊòØPythonÁöÑ‰∏ÄÂÄãÁπ™ÂúñÂ∫´
from sklearn.tree import plot_tree #plot_treeÂáΩÊï∏ÂèØ‰ª•Áî®‰æÜÁπ™Ë£ΩÊ±∫Á≠ñÊ®πÁöÑÂúñÂΩ¢Ë°®Á§∫
plt.figure(figsize=(10,10))
plot_tree(reg,
     feature_names=diabetes["feature_names"], #4Á®ÆÁâπÂæµ=feature_names
     filled=True)
```

```python
#Â∞çÁ≠îÊ°à: Ë©ïÂàÜÊñπÂºèÊòØÊúâÂæàÂ§öÁ®Æ, Ê≠£Á¢∫Áéá(accuracy)
#Ë°°Èáè(x_test„ÄÅy_test)>preiÔºõ(1/n)(È†êÊ∏¨ÂÄºprei-Ê≠£Á¢∫Á≠îÊ°àytesti)**2
from sklearn.metrics import r2_score
pre=reg.predict(x_test)
r2_score(y_test, pre)

#max_depth=2>0.33Ôºõ=3>0.35Ôºõ=4>0.37Ôºõ=5>0.25Ôºõ=6>0.21Ôºõ=7>0.20Ôºõ=8>-0.03Ôºõ=9>-0.23
```

```python
#‰∫§ÂèâÈ©óË≠âÔºàcross-validationÔºâ‰æÜË©ï‰º∞ DecisionTreeRegressor Ê±∫Á≠ñÊ®πËø¥Ê≠∏Ê®°ÂûãÁöÑÈ†êÊ∏¨ËÉΩÂäõ
from sklearn.model_selection import cross_val_score
reg=DecisionTreeRegressor(max_depth=4)
scores=cross_val_score(reg,x,y, scoring='r2', cv=10, n_jobs=1)
print('10:,',scores)
print('avg:,',np.average(scores))
#max_depth=2>0.30Ôºõ=3>0.28Ôºõ=4>0.27
```

```python
from sklearn.manifold import  TSNE # t-SNE (t-distributed Stochastic Neighbor Embedding) ÁÆóÊ≥ï‰æÜË¶ñË¶∫ÂåñÈ´òÁ∂≠Êï∏ÊìöÈõÜ
from sklearn.preprocessing import MinMaxScaler

from sklearn.datasets import load_iris

data = load_iris()
scaler = MinMaxScaler() #ÈÄ≤Ë°åÊï∏ÊìöÁöÑÊ≠£Ë¶èÂåñ„ÄÇÂÆÉÊúÉÊääÊï∏ÊìöÁ∏ÆÊîæÂà∞‰∏ÄÂÄãÁµ¶ÂÆöÁöÑÊúÄÂ∞èÂÄºÂíåÊúÄÂ§ßÂÄºÔºàÈÄöÂ∏∏ÊòØ 0-1Ôºâ‰πãÈñì
data_scale = scaler.fit_transform(data["data"]) # Â∞ç iris Êï∏ÊìöÈõÜÁöÑÁâπÂæµÊï∏ÊìöÈÄ≤Ë°åÁ∏ÆÊîæÔºåÂ∞áÊâÄÊúâÁâπÂæµÁ∏ÆÊîæÂà∞ 0 Âíå 1 ‰πãÈñì
tsne = TSNE(n_components=2, n_jobs=-1) #ÂâµÂª∫‰∫Ü‰∏ÄÂÄã t-SNE Áâ©‰ª∂ÔºåÊ∫ñÂÇôÂ∞áÊï∏ÊìöÁ∏ÆÊ∏õÂà∞ 2 Á∂≠ #‰ΩøÁî®ÊâÄÊúâÁöÑ CPU Ê†∏ÂøÉÈÄ≤Ë°åË®àÁÆó
data_tsne = tsne.fit_transform(data_scale) #‰ΩøÁî® t-SNE Â∞çÊ≠£Ë¶èÂåñÂæåÁöÑ iris Êï∏ÊìöÈõÜÈÄ≤Ë°åÈôçÁ∂≠
data_tsne
```

```python
import matplotlib.pyplot as plt #‰ΩøÁî® matplotlib Âíå seaborn ÂÖ©ÂÄã Python ÁöÑÊï∏ÊìöË¶ñË¶∫ÂåñÂ∫´‰æÜÁπ™Ë£ΩÊï£ÈªûÂúñ
import seaborn as sns
#Âæû data_tsneÔºà‰∏ÄÂÄã‰∫åÁ∂≠Èô£ÂàóÔºåÈÄöÂ∏∏ÊòØÊ©üÂô®Â≠∏ÁøíÊ®°ÂûãÁöÑËº∏Âá∫Ôºâ‰∏≠ÂàÜÂà•ÂèñÂá∫ÊâÄÊúâË°åÁöÑÁ¨¨‰∏ÄÂàóÂíåÁ¨¨‰∫åÂàóÊï∏ÊìöÔºå‰∏¶ÂàÜÂà•Ë≥¶ÂÄºÁµ¶ xs Âíå ys
xs = data_tsne[:, 0] #ÊâÄÊúâÊ®£Êú¨ÁöÑÁ¨¨‰∏ÄÂÄãÁâπÂæµ
ys = data_tsne[:, 1] #ÊâÄÊúâÊ®£Êú¨ÁöÑÁ¨¨‰∫åÂÄãÁâπÂæµ
sns.scatterplot(x=xs,
        y=ys,
        hue=data["target"]) #hue=data["target"] ÊåáÂÆö‰∫ÜÊï£ÈªûÁöÑÈ°èËâ≤ÔºåÈ°èËâ≤Áî± data["target"]ÔºàÊ®£Êú¨ÁöÑÊ®ôÁ±§ÔºâÊ±∫ÂÆö
```

```python
from sklearn.datasets import load_wine
data = load_wine()
scaler = MinMaxScaler()
data_scale = scaler.fit_transform(data["data"])
tsne = TSNE(n_components=2, n_jobs=-1)
data_tsne = tsne.fit_transform(data_scale)
xs = data_tsne[:, 0]
ys = data_tsne[:, 1]
sns.scatterplot(x=xs,
                y=ys,
                hue=data["target"])
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
scaler = MinMaxScaler()
data_scale = scaler.fit_transform(data["data"])
tsne = TSNE(n_components=2, n_jobs=-1)
data_tsne = tsne.fit_transform(data_scale)
xs = data_tsne[:, 0]
ys = data_tsne[:, 1]
sns.scatterplot(x=xs,
        y=ys,
        hue=data["target"])
```
