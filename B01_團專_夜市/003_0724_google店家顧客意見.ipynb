```python
df['評論內容'] = df['評論內容'].replace('\n', '', regex=True)
df.head(20)
```

```python
import re

df['評論內容'] = df['評論內容'].apply(lambda x: re.sub(r'[^\w\s$]', '', x)) #去除奇怪表情符號，但保留$
df.head(20)
```

```python
#匯入必要套件
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import jieba
import jieba.analyse
from collections import Counter # 次數統計
```

```python
def commentcut(p):
    return " ".join(jieba.cut(p))
comment_list = df['評論內容'].apply(commentcut)
comment_list
```

```python
# 將評論內容合併成一個字串
text = ' '.join(df['評論內容'].dropna())

# 使用jieba進行分詞
seg_list = jieba.cut(text)
seg_list
# 將分詞結果轉為空格分隔的字串
seg_text = ' '.join(seg_list)
seg_text
```

```python
# 打开一个文件，如果文件不存在，那么就创建一个新的文件
file = open("words.txt", "w")
# 将字符串写入到文件中
file.write(seg_text)
# 关闭文件
file.close()
```

```python
#匯入相關中文字典檔
dictfile = "dict.txt"  # 上傳dict.txt字典檔  #修改1
stopfile = "stopwords.txt"  # 上傳stopwordstxt檔  #修改2
fontpath = "msjh.ttc"  # 上傳繁體中文字型檔 #修改3
mdfile = "words.txt"  # 上傳主要中文文檔chinesewords.txt #修改4
pngfile = "circle.png"  # 剛才下載存的底圖 #修改5
alice_mask = np.array(Image.open(pngfile))
```

```python
jieba.set_dictionary(dictfile)
jieba.analyse.set_stop_words(stopfile)

text = open(mdfile,"r",encoding="utf-8").read()

tags = jieba.analyse.extract_tags(text, topK=25)

seg_list = jieba.lcut(text, cut_all=False)
dictionary = Counter(seg_list)

freq = {}
for ele in dictionary:
    if ele in tags:
        freq[ele] = dictionary[ele]
print(freq) # 計算出現的次數

#繪圖
wordcloud = WordCloud(width = 4000, height = 3000,background_color="white", mask=alice_mask, contour_width=3, contour_color='steelblue', font_path= fontpath,max_font_size=300,min_font_size=10,colormap = "Set2").generate_from_frequencies(freq)
plt.figure(figsize=(25,15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()

import datetime
#儲存檔案
date = str(datetime.date.today())
wordcloud.to_file(date+'_output.png')
```
