```python
#åŒ¯å…¥å¿…è¦å¥—ä»¶
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import jieba
import jieba.analyse
from collections import Counter # æ¬¡æ•¸çµ±è¨ˆ
```

```python
def commentcut(p):
    return " ".join(jieba.cut(p))
comment_list = df['è©•è«–å…§å®¹'].apply(commentcut)
comment_list
```

```python
# å°‡è©•è«–å…§å®¹åˆä½µæˆä¸€å€‹å­—ä¸²
text = ' '.join(df['è©•è«–å…§å®¹'].dropna())

# ä½¿ç”¨jiebaé€²è¡Œåˆ†è©
seg_list = jieba.cut(text)
seg_list
# å°‡åˆ†è©çµæœè½‰ç‚ºç©ºæ ¼åˆ†éš”çš„å­—ä¸²
seg_text = ' '.join(seg_list)
seg_text
```

```python
# æ‰“å¼€ä¸€ä¸ªæ–‡ä»¶ï¼Œå¦‚æœæ–‡ä»¶ä¸å­˜åœ¨ï¼Œé‚£ä¹ˆå°±åˆ›å»ºä¸€ä¸ªæ–°çš„æ–‡ä»¶
file = open("words.txt", "w")
# å°†å­—ç¬¦ä¸²å†™å…¥åˆ°æ–‡ä»¶ä¸­
file.write(seg_text)
# å…³é—­æ–‡ä»¶
file.close()
```

```python
#åŒ¯å…¥ç›¸é—œä¸­æ–‡å­—å…¸æª”
dictfile = "dict.txt"  # ä¸Šå‚³dict.txtå­—å…¸æª”  #ä¿®æ”¹1
stopfile = "stopwords.txt"  # ä¸Šå‚³stopwordstxtæª”  #ä¿®æ”¹2
fontpath = "msjh.ttc"  # ä¸Šå‚³ç¹é«”ä¸­æ–‡å­—å‹æª” #ä¿®æ”¹3
mdfile = "words.txt"  # ä¸Šå‚³ä¸»è¦ä¸­æ–‡æ–‡æª”chinesewords.txt #ä¿®æ”¹4
pngfile = "circle.png"  # å‰›æ‰ä¸‹è¼‰å­˜çš„åº•åœ– #ä¿®æ”¹5
alice_mask = np.array(Image.open(pngfile))
```

```python
jieba.set_dictionary(dictfile)
jieba.analyse.set_stop_words(stopfile)

text = open(mdfile,"r",encoding="utf-8").read()

tags = jieba.analyse.extract_tags(text, topK=25)

seg_list = jieba.lcut(text, cut_all=False)
dictionary = Counter(seg_list)

freq = {}
for ele in dictionary:
    if ele in tags:
        freq[ele] = dictionary[ele]
print(freq) # è¨ˆç®—å‡ºç¾çš„æ¬¡æ•¸

#ç¹ªåœ–
wordcloud = WordCloud(width = 4000, height = 3000,background_color="white", mask=alice_mask, contour_width=3, contour_color='steelblue', font_path= fontpath,max_font_size=300,min_font_size=10,colormap = "Set2").generate_from_frequencies(freq)
plt.figure(figsize=(25,15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()

import datetime
#å„²å­˜æª”æ¡ˆ
date = str(datetime.date.today())
wordcloud.to_file(date+'_output.png')
```
